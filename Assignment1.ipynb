(g) Multiply the following matrices and also find covariance between matrices using NumPy:

i. p = [[1, 2], [2, 3]]
q = [[4, 5], [6, 7]]

```python
import numpy as np
p1 = np.array([[1, 2], [2, 3]])
q1 = np.array([[4, 5], [6, 7]])
product1 = np.matmul(p1, q1)
cov1 = np.cov(p1.flatten(), q1.flatten())
print("i. Matrix Multiplication:\n", product1)
print("i. Covariance matrix:\n", cov1)
```

(h) For the matrices: x = np.array([[2, 3, 4], [3, 2, 9]]); y = np.array([[1, 5, 0], [5, 10, 3]]),
find inner, outer and cartesian product?

```python
import numpy as np
x = np.array([[2, 3, 4], [3, 2, 9]])
y = np.array([[1, 5, 0], [5, 10, 3]])
inner_product = np.inner(x, y)
print("Inner product:\n", inner_product)
outer_product = np.outer(x, y)
print("Outer product:\n", outer_product)
from itertools import product
cartesian_product = np.array([np.concatenate((a, b)) for a, b in product(x, y)])
print("Cartesian product (row-wise combinations):\n", cartesian_product)
```

(a) For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])
i. Find element-wise absolute value
ii. Find the 25th, 50th, and 75th percentile of flattened array, for each column, for each
row.
iii. Mean, Median and Standard Deviation of flattened array, of each column, and
each row

```python
import numpy as np
array = np.array([[1, -2, 3],
                  [-4, 5, -6]])
abs_array = np.abs(array)
print("i. Absolute values:\n", abs_array)

percentiles_flat = np.percentile(array, [25, 50, 75])
percentiles_cols = np.percentile(array, [25, 50, 75], axis=0)
percentiles_rows = np.percentile(array, [25, 50, 75], axis=1)
print("ii. Percentiles of flattened array (25th, 50th, 75th):", percentiles_flat)
print("\nPercentiles per column:\n", percentiles_cols)
print("\nPercentiles per row:\n", percentiles_rows)

mean_flat = np.mean(array)
median_flat = np.median(array)
std_flat = np.std(array)
mean_col = np.mean(array, axis=0)
median_col = np.median(array, axis=0)
std_col = np.std(array, axis=0)
mean_row = np.mean(array, axis=1)
median_row = np.median(array, axis=1)
std_row = np.std(array, axis=1)

print("iii. Mean (flattened):", mean_flat)
print("Median (flattened):", median_flat)
print("Std Dev (flattened):", std_flat)
print("\nMean per column:", mean_col)
print("Median per column:", median_col)
print("Std Dev per column:", std_col)
print("\nMean per row:", mean_row)
print("Median per row:", median_row)
print("Std Dev per row:", std_row)
```

(a) Write a function named img_to_array(path) that reads an image from a specified path and
save it as text file on local machine? (Note: use separate cases for RGB and Grey Scale
images)

```python
import cv2
import numpy as np

def img_to_array(path, save_as="image_data.txt"):
    # Try reading as a grayscale image
    img_gray = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    if img_gray is not None:
        print("Image is in Grayscale")
        np.savetxt(save_as, img_gray, fmt='%d')
        return img_gray
    
    # Try reading as a color image
    img_rgb = cv2.imread(path, cv2.IMREAD_COLOR)
    if img_rgb is not None:
        print("Image is in RGB")
        # Convert from BGR to RGB
        img_rgb = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2RGB)
        # Save each channel separately
        with open(save_as, 'w') as f:
            for channel, name in zip(cv2.split(img_rgb), ['Red', 'Green', 'Blue']):
                f.write(f"--- {name} Channel ---\n")
                np.savetxt(f, channel, fmt='%d')
                f.write("\n")
        return img_rgb

    raise ValueError("Image not found or unsupported format.")
```
